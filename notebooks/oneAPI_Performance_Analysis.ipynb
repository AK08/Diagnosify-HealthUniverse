## oneAPI Performance Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Alzheimer's Disease Prediction

```python

# Sample data, replace with your actual data
system_names = ['Without oneAPI', 'With oneAPI']
execution_times = [4283.03, 695.42]  # Replace with actual execution times

# Calculate percentage decrease in time
time_decrease = ((execution_times[0] - execution_times[1]) / execution_times[0]) * 100

sns.set_palette("pastel")

fig, axs = plt.subplots(figsize=(6, 5))

# Execution Time Comparison
axs.bar(system_names, execution_times, color=['#004a86', '#00c7fd'])
axs.set_ylabel('Execution Time (seconds)')
axs.set_title('Execution Time Comparison\nTime Decrease: {:.2f}%'.format(time_decrease))

plt.tight_layout()
plt.show()
```
